```python
# LOADING THE DATASET

import pandas as pd

df = pd.read_csv("student-mat.csv", sep=';')  # Data is separated by ;
df.head() #first 5 rows
```

```python
#DATA EXPLORATION

df.isnull().sum() # empty cells
df.dtypes #data types
df.shape #rows and columns
```

```python
# DATA CLEANING

df = df.dropna() # Removes rows with missing values (if any)
df = df.drop_duplicates() # Remove duplicate rows
```

```python
#Data Analysis Questions

# 1. What is the average final grade (G3)?

average_g3 = df['G3'].mean()
print("Average Final Grade (G3):", average_g3)
```

```python
# 2. How many students scored above 15 in G3?

high_scores = df[df['G3'] > 15]
print("Number of students scoring above 15:", high_scores.shape[0])
```

```python
# 3. Is there a correlation between study time and final grade?

correlation = df['studytime'].corr(df['G3'])
print("Correlation between study time and final grade (G3):", correlation)
```

```python
# 4. Which gender has a higher average G3?

avg_by_gender = df.groupby('sex')['G3'].mean()
print("Average G3 by gender:")
print(avg_by_gender)
```

```python
#DATA VISUALIZATION

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
#1. Histogram of final grades (G3)

plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.show()
```

This histogram shows how the students' final exam scores (G3) are distributed.

Most students scored around 10 to 12, indicating the average performance level.

```python
#2. Scatter plot between study time and final grade

plt.figure(figsize=(8,5))
plt.scatter(df['studytime'], df['G3'], color='green')
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time (1 = <2hrs, 4 = >10hrs)')
plt.ylabel('Final Grade (G3)')
plt.show()
```

This scatter plot shows the relationship between weekly study time and final exam grades.

Each dot is a student.

```python
#3. Bar chart comparing male vs female average G3

avg_by_gender.plot(kind='bar', color=['purple', 'orange'])
plt.title('Average Final Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average G3')
plt.xticks(rotation=0)
plt.show()
```

This bar chart compares the average final exam scores between male and female students.

From the chart, male students have a slightly higher average score compared to female students.
